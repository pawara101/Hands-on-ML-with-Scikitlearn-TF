# Desicion Tree Classifier

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = load_iris()
```

```python
X = iris['data'][:,2:] #Petal Length & Width
y = iris['target']
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)
```
